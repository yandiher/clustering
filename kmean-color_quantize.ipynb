```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_as_array = mpimg.imread('palm-trees.jpg')
```

```python
plt.imshow(image_as_array);
```

```python
h, w, c = image_as_array.shape
```

```python
image_as_array2D = image_as_array.reshape(h*w, c)
```

```python
print(len(image_as_array.shape))
print(len(image_as_array2D.shape))
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, n_init='auto')
label = model.fit_predict(image_as_array2D)
```

```python
rgb = model.cluster_centers_.round().astype(int)
```

```python
rgb
```

```python
quantized_image = np.reshape(rgb[label],(h, w, c))
```

```python
quantized_image.shape
```

```python
plt.imshow(quantized_image)
```
