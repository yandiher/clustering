# import basic library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
def display(w=16, h=5):
    plt.figure(figsize=(w,h), dpi=200)
```

# load the data

```python
df = pd.read_csv('cancer-tumor-data-features.csv')
df.head()
```

# preprocessing the data manually

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scale = scaler.fit_transform(df)
```

```python
cov_matrix = np.cov(x_scale, rowvar=False)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)
```

```python
num_components = 2
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values = eigen_values[sorted_key]
eigen_vectors = eigen_vectors[:,sorted_key]
```

```python
principal_components = np.dot(x_scale, eigen_vectors)
```

```python
display()
sns.scatterplot(x=principal_components[:,0], y=principal_components[:,1])
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
display()
sns.scatterplot(x=principal_components[:,0], y=principal_components[:,1], hue=cancer['target'])
plt.show()
```

# preprocessing data with scikit-learn

```python
x = df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scale = scaler.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
result = pca.fit_transform(x_scale)
```

```python
display()
sns.scatterplot(x=result[:,0], y=result[:,1])
plt.show()
```

```python
explained_variance = []
number_of_components = range(1,30)

for n in number_of_components:
    pca_evaluate = PCA(n_components=n)
    pca_evaluate.fit(x_scale)
    components = np.sum(pca_evaluate.explained_variance_ratio_)
    explained_variance.append(components)
```

```python
display()
sns.lineplot(x=number_of_components, y=explained_variance)
plt.xlabel('number of components')
plt.ylabel('score')
plt.show()
```
