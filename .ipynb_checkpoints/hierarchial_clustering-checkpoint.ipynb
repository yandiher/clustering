# import basic library

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def adjust(w=16,h=5):
    plt.figure(figsize=(w, h), dpi=200)
```

# load the data

```python
df = pd.read_csv('cluster-mpg.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.describe().round(2)
```

```python
df.origin.value_counts()
```

there are 3 unique values for origin.

```python
df.name.nunique()
```

there are 301 unique values for name.

```python
x = df.drop(['name'], axis=1)
x = pd.get_dummies(data=x)
x.head()
```

# preprocessing data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scale = scaler.fit_transform(x)
```

```python
x_scale = pd.DataFrame(data=x_scale, columns=x.columns)
```

```python
sns.clustermap(x_scale, row_cluster=False)
plt.show()
```

```python
sns.clustermap(x_scale, col_cluster=False)
plt.show()
```

# build the model

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4)
```

```python
label = model.fit_predict(x_scale)
```

```python
len(label), len(x_scale)
```

```python
adjust()
sns.scatterplot(data=x, x='mpg', y='weight', hue=label, palette='inferno')
plt.show()
```
