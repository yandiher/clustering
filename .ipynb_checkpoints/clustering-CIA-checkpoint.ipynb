```python
data[data['Agriculture'].isna()]['Country']
```

```python
data[data['Agriculture'].isna()] = data[data['Agriculture'].isna()].fillna(0)
```

```python
data.isna().sum()
```

```python
data['Climate'] = data['Climate'].fillna(
    data.groupby(['Region'])['Climate'].transform('mean')
)
```

```python
data['Literacy (%)'] = data['Literacy (%)'].fillna(
    data.groupby(['Region'])['Literacy (%)'].transform('mean')
)
```

```python
data = data.dropna()
```

```python
data.isna().sum()
```

```python
x = data.drop(['Country'], axis=1)
x = pd.get_dummies(data=x)
x.head()
```

# build the model

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scale = scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans

number = np.arange(2, 31)
ssd = []

for n in number:
    model = KMeans(n_clusters=n, n_init='auto')
    model.fit(x_scale)
    ssd.append(model.inertia_)
```

```python
fix()
plt.plot(number, ssd, 'o--')
plt.show()
```

```python
fix()
diff = pd.Series(ssd).diff().plot(kind='bar')
plt.show()
```

```python
model = KMeans(n_clusters=4, n_init='auto')
model.fit(x_scale)
```

```python
x['cluster'] = model.labels_
x.head()
```
